# Feature Engineering

```python
df["TotArea"] = df["TotalBsmtSF"] + df["1stFlrSF"] + df["2ndFlrSF"]
```

```python
object_att = df.dtypes[df.dtypes == "object"]
```

```python
numeric_att = df.dtypes[df.dtypes != "object"].index
```

```python
skew = df[numeric_att].apply(lambda x: x.skew()).sort_values(ascending=False)
skewness = pd.DataFrame(skew).dropna()
```

```python
skewness = skewness[abs(skewness)>0.5]
```

```python
skewness.dropna(inplace=True)
```

```python
skewness
```

```python
from scipy.special import boxcox1p
```

```python
for entry in skewness.index:
    df[entry] = boxcox1p(df[entry], 0.5)
```
